```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
logs = np.load("history3.npy", allow_pickle=True)
```

```python
testing_loss = np.zeros((9, 4))

x_counter = 0
y_counter = 0
for i in range(len(logs)):
    testing_loss[y_counter][x_counter % 4] = logs[i]['logs']['testing_loss']
    x_counter += 1
    if x_counter % 4 == 0:
        y_counter += 1
        
plt.imshow(testing_loss)
```

```python
testing_acc = np.zeros((9, 4))

x_counter = 0
y_counter = 0
for i in range(len(logs)):
    testing_acc[y_counter][x_counter % 4] = logs[i]['logs']['testing_acc']
    x_counter += 1
    if x_counter % 4 == 0:
        y_counter += 1
        
plt.imshow(testing_acc)
```
