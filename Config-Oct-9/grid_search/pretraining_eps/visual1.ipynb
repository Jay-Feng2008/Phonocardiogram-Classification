```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
logs = np.load("history1.npy", allow_pickle=True)
```

```python
testing_loss = np.zeros((5, 4))

x_counter = 0
y_counter = 0
for i in range(len(logs)):
    testing_loss[y_counter][x_counter % 4] = logs[i]['logs']['testing_loss']
    x_counter += 1
    if x_counter % 4 == 0:
        y_counter += 1
        
plt.imshow(testing_loss)
```

```python
testing_acc = np.zeros((5, 4))

x_counter = 0
y_counter = 0
for i in range(len(logs)):
    testing_acc[y_counter][x_counter % 4] = logs[i]['logs']['testing_acc']
    x_counter += 1
    if x_counter % 4 == 0:
        y_counter += 1
        
plt.imshow(testing_acc)
```

```python
training_loss = np.zeros((5, 4))

x_counter = 0
y_counter = 0
for i in range(len(logs)):
    training_loss[y_counter][x_counter % 4] = logs[i]['logs']['training_loss']
    x_counter += 1
    if x_counter % 4 == 0:
        y_counter += 1
        
plt.imshow(training_loss)
```

```python
training_acc = np.zeros((5, 4))

x_counter = 0
y_counter = 0
for i in range(len(logs)):
    training_acc[y_counter][x_counter % 4] = logs[i]['logs']['training_acc']
    x_counter += 1
    if x_counter % 4 == 0:
        y_counter += 1
        
plt.imshow(training_acc)
```

From the plots above, we can see that the optimal pretraining steps are in [2, 3, 4], the eps are in [10, 20, 30]
